```python
selF = ['LSTAT', 'RM']
x_train, x_test, y_train, y_test = train_test_split(df[selF], df[[target]], test_size=0.2, random_state=0)
```

```python
alphaVals = np.arange(0.01, 5, 0.01)
alphaVals.shape
```

```python
testVals = []
trainVals = []

for val in alphaVals:
    model = Ridge(alpha=val)
    model.fit(x_train, y_train)
    testVals.append(model.score(x_test, y_test) )
    trainVals.append(model.score(x_train, y_train))
np.argmax(testVals), alphaVals[np.argmax(testVals)], np.max(testVals), np.argmax(trainVals), np.max(trainVals)
```

```python
def plot_fig(model, x_train, x_test, y_train, y_test):
    plt.scatter(x_train, y_train, color='blue')
    plt.scatter(x_test, y_test, color='red')
    plt.plot(x_test, model.predict(x_test))
    plt.show()
```

```python
plot_fig(model, x_train, x_test, y_train, y_test)
```

```python
alphaVals = np.arange(0.01, 100, 0.01)
alphaVals.shape
```

```python
testVals = []
trainVals = []
models = []
for val in alphaVals:
    model = Ridge(alpha=val)
    model.fit(x_train, y_train)
    testVals.append(model.score(x_test, y_test) )
    trainVals.append(model.score(x_train, y_train))
    models.append(model)
np.argmax(testVals), alphaVals[np.argmax(testVals)], np.max(testVals), np.argmax(trainVals), np.max(trainVals)
```

```python
plt.plot(alphaVals, testVals, color='red')
plt.plot(alphaVals, trainVals)
plt.show()
```

```python
plot_fig(models[np.argmax(testVals)], x_train, x_test, y_train, y_test)
```

```python
bestModel = models[np.argmax(testVals)]
```

```python
y_train_pred = bestModel.predict(x_train)
y_test_pred = bestModel.predict(x_test)
plt.scatter(y_train_pred, (y_train_pred - y_train), color='blue' )
plt.scatter(y_test_pred, (y_test_pred - y_test), color='red' )
plt.hlines(0, -10, 50)
plt.xlim([-10,50])
plt.show()
```
