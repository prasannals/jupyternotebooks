```python
from sklearn import datasets
from sklearn.cross_validation import KFold
from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import cross_val_score
```

```python
boston = datasets.load_boston()
```

```python
boston_X = boston.data
boston_y = boston.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(boston_X, boston_y)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```

```python
kf = KFold(len(x_train), n_folds=10)
models = []
for train, test in kf:
    tempReg = LinearRegression()
    tempReg.fit(x_train[train], y_train[train])
    print(tempReg.score(x_train[test], y_train[test]))
    models.append(tempReg)
```

```python
models[3].score(x_test, y_test)
```

```python
scores = cross_val_score(estimator=reg, X=x_train, y= y_train, cv=KFold(len(x_train), n_folds=10), n_jobs=4)
```

```python
scores = []
for i in range(5, 50):
    sc = cross_val_score(estimator=reg, X=x_train, y= y_train, cv=KFold(len(x_train), n_folds=i), n_jobs=4)
    #plt.plot(range(i), sc)
    #plt.ylim((0,1))
    #plt.show()
    scores.append(sc)
```

```python
score_var = [np.var(score) for score in scores]
```

```python
plt.plot(range(len(score_var)), score_var)
```

```python
best_score = [np.max(score) for score in scores]
```

```python
plt.plot(range(len(best_score)), best_score)
```

```python
residue = [(1 - score) for score in best_score]
```

```python
plt.plot(range(len(residue)), residue)
```
